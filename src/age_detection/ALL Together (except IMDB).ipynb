```python
import os
import pandas as pd

# Function to extract age from the last two digits of the file name
def extract_age_from_filename(filename):
    try:
        age = int(filename[-6:-4])  # Extract the last two digits of the file name
        return age
    except ValueError:
        return None

# Function to determine gender based on serial number
def determine_gender(person_id):
    if 0 <= person_id <= 7380:
        return "female"
    elif 7381 <= person_id <= 13321:
        return "male"
    else:
        return None

# Path to the folder containing photos
folder_path = "C:/Users/HyperWorks/Downloads/All-Age-Faces_Dataset (ASIANS)/"

# Create empty lists to store data
file_names = []
ages = []
genders = []

# Iterate through the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".jpg"):
        file_names.append(filename)  # Store only the file name
        age = extract_age_from_filename(filename)
        person_id = int(filename[:5])  # Extract the first 5 digits as person_id
        gender = determine_gender(person_id)
        
        if age is not None:
            ages.append(age)
            genders.append(gender)
        else:
            # Handle the case when the age couldn't be extracted
            print(f"Ignoring file with an invalid age format: {filename}")

# Create a DataFrame
data = {
    "file_name": file_names,
    "age": ages,
    "gender": genders,
}
df = pd.DataFrame(data)
df.insert(2, 'name', None)

# Display the DataFrame
print(df)
```

```python
import os
import pandas as pd

def extract_age_and_gender(filename):
    parts = filename.split('_')
    if len(parts) >= 2:
        age = int(parts[0])
        gender = parts[1]
        # Map '0' to 'male' and '1' to 'female'
        gender = 'male' if gender == '0' else 'female' if gender == '1' else None
        return age, gender
    else:
        return None, None

def create_dataframe_from_photos(folder_path):
    data = []

    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg"):
            age, gender = extract_age_and_gender(filename)
            if age is not None and gender is not None:
                data.append([filename, age, gender])

    if data:
        df = pd.DataFrame(data, columns=["file_name", "age", "gender"])
        return df
    else:
        return None

```

### KANFace

```python
import os
import pandas as pd
import re

# Define a function to extract age from file names using regular expressions
def extract_age_from_filename(filename):
    pattern = r'\_(\d{2})_'
    match = re.search(pattern, filename)
    if match:
        age = int(match.group(1))
        return age
    return None

# Specify the folder containing the photos
folder_path = 'C:/Users/HyperWorks/Downloads/images'

# Initialize lists to store file paths and ages
file_paths = []
ages = []

# Iterate through the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.png'):  # Check if the file is a PNG file
        age = extract_age_from_filename(filename)
        if age is not None:
            file_paths.append(filename)
            ages.append(age)

# Create a DataFrame from the collected data
data = {'file_name': file_paths, 'age': ages}
df = pd.DataFrame(data)

df.insert(2, 'name', None)
df.insert(3, 'gender', None)

# Display the DataFrame
print(df)
```

```python
import os
import pandas as pd

def extract_age_gender_from_filename(filename):
    """
    Extracts age and gender from the given filename.
    Example filename: '16487_SerenaWilliams_33_f.jpg'
    """
    parts = filename.split('_')
    if len(parts) >= 3 and parts[-1][0] in ['f', 'm']:
        age = parts[-2]
        gender = 'female' if parts[-1][0] == 'f' else 'male'
        return age, gender
    return None, None

def process_directory(directory_path):
    data = []
    for filename in os.listdir(directory_path):
        if filename.endswith('.jpg'):
            age, gender = extract_age_gender_from_filename(filename)
            if age and gender:
                data.append({'file_name': filename, 'age': age, 'gender': gender})

    df = pd.DataFrame(data)
    return df

directory_path = 'C:/Users/HyperWorks/Downloads/AgeDB'
df = process_directory(directory_path)
df.insert(2, 'name', None)
```
