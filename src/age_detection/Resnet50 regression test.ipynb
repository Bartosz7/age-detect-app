```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Assuming you have a dataset in 'dataset_dir' directory
transform = transforms.Compose([transforms.Resize((224, 224)),
                                transforms.ToTensor()])
dataset = datasets.ImageFolder(root='/home/simonexc/Documents/age-detect-app/data/train_data', transform=transform)
dataloader = DataLoader(dataset, batch_size=256, shuffle=False)

def compute_mean_std(dataloader):
    channels_sum, channels_squared_sum, num_batches = 0, 0, 0

    for data, _ in dataloader:
        channels_sum += torch.mean(data, dim=[0, 2, 3])
        channels_squared_sum += torch.mean(data ** 2, dim=[0, 2, 3])
        num_batches += 1

    mean = channels_sum / num_batches
    std = (channels_squared_sum / num_batches - mean ** 2) ** 0.5

    return mean, std

mean, std = compute_mean_std(dataloader)
print(f'Mean: {mean}')
print(f'Std: {std}')
```

```python
import os
import pyheif
from PIL import Image

def read_heic(file_path):
    heif_file = pyheif.read(file_path)
    return Image.frombytes(
        heif_file.mode, 
        heif_file.size, 
        heif_file.data,
        "raw",
        heif_file.mode,
        heif_file.stride,
    )

def convert_heic_to_jpg(directory):
    for filename in os.listdir(directory):
        if filename.lower().endswith(".heic"):
            file_path = os.path.join(directory, filename)
            try:
                img = read_heic(file_path)
                jpg_path = os.path.join(directory, filename.replace(".HEIC", ".jpg").replace(".heic", ".jpg"))
                img.save(jpg_path, "JPEG")
                print(f"Converted {filename} to JPG")
            except Exception as e:
                print(f"Failed to convert {filename}: {e}")

# Replace 'your_directory_path' with the path to the folder containing HEIC files
convert_heic_to_jpg('/home/simonexc/Documents/age-detect-app/data/test_data/face_detection')
```
